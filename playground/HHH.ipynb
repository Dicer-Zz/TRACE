```python
from matplotlib import pyplot as plt
import numpy as np

# instruction-follow dataset: lima
# safty dataset: cona

# 分为左右两个子图：左边是指令遵循能力测试（LIMA），右边是安全性测试（CoNA）
# 每个子图包含三个实验：EPI vs. Base, O-LoRA vs. Base, EPI vs. O-LoRA
# 每个实验包含两个结果：AI, HUMAN，使用堆叠柱状图表示（需首先归一化为百分数）


# 1. EPI vs. Base
# 2. O-LoRA vs. Base
# 3. EPI vs. O-LoRA

# define subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
# 深蓝、中蓝、浅蓝
colors = ["#1f77b4", "#aec7e8", "#dce7f6"]
color_labels = ["Win", "Tie", "Lose"]

# 使用横向的堆叠柱状图表示
# EPI vs. Base
# LIMA
# AI & Human

ai_score = HHH_score["EPI vs. Base"]["LIMA"]["AI"]
human_score = HHH_score["EPI vs. Base"]["LIMA"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

bar_width = 0.8
for i in range(3):
    axs[0].bar(1, ai_score[2-i], bar_width, color=colors[2-i], label=color_labels[2-i])
    axs[0].bar(2, human_score[2-i], bar_width, color=colors[2-i])

# O-LoRA vs. Base
# LIMA
# AI & Human
ai_score = HHH_score["O-LoRA vs. Base"]["LIMA"]["AI"]
human_score = HHH_score["O-LoRA vs. Base"]["LIMA"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

for i in range(3):
    axs[0].bar(4, ai_score[2-i], bar_width, color=colors[2-i])
    axs[0].bar(5, human_score[2-i], bar_width, color=colors[2-i])

# EPI vs. O-LoRA
# LIMA
# AI & Human
ai_score = HHH_score["EPI vs. O-LoRA"]["LIMA"]["AI"]
human_score = HHH_score["EPI vs. O-LoRA"]["LIMA"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

for i in range(3):
    axs[0].bar(7, ai_score[2-i], bar_width, color=colors[2-i])
    axs[0].bar(8, human_score[2-i], bar_width, color=colors[2-i])

# 使用横向的堆叠柱状图表示
# EPI vs. Base
# LIMA
# AI & Human

ai_score = HHH_score["EPI vs. Base"]["CoNa"]["AI"]
human_score = HHH_score["EPI vs. Base"]["CoNa"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

bar_width = 0.8
for i in range(3):
    axs[1].bar(1, ai_score[2-i], bar_width, color=colors[2-i], label=color_labels[2-i])
    axs[1].bar(2, human_score[2-i], bar_width, color=colors[2-i])

# O-LoRA vs. Base
# CoNa
# AI & Human
ai_score = HHH_score["O-LoRA vs. Base"]["CoNa"]["AI"]
human_score = HHH_score["O-LoRA vs. Base"]["CoNa"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

for i in range(3):
    axs[1].bar(4, ai_score[2-i], bar_width, color=colors[2-i])
    axs[1].bar(5, human_score[2-i], bar_width, color=colors[2-i])

# EPI vs. O-LoRA
# CoNa
# AI & Human
ai_score = HHH_score["EPI vs. O-LoRA"]["CoNa"]["AI"]
human_score = HHH_score["EPI vs. O-LoRA"]["CoNa"]["HUMAN"]
# normalize
ai_score = np.array(ai_score) / np.sum(ai_score)
human_score = np.array(human_score) / np.sum(human_score)

# accumlate the score
ai_score = np.cumsum(ai_score)
human_score = np.cumsum(human_score)

for i in range(3):
    axs[1].bar(7, ai_score[2-i], bar_width, color=colors[2-i])
    axs[1].bar(8, human_score[2-i], bar_width, color=colors[2-i])

# 将x轴替换为实验名称 ["EPI vs. Base", "O-LoRA vs. Base", "EPI vs. O-LoRA"]
axs[0].set_xticks([1.5, 4.5, 7.5])
axs[0].set_xticklabels(["EPI vs. Base", "O-LoRA vs. Base", "EPI vs. O-LoRA"])
axs[0].set_ylabel("% Win Rate")
axs[0].set_title("LIMA")
# 添加颜色的含义图例 color for Win, Tie, Lose
axs[0].legend(loc="upper right")

axs[1].set_xticks([1.5, 4.5, 7.5])
axs[1].set_xticklabels(["EPI vs. Base", "O-LoRA vs. Base", "EPI vs. O-LoRA"])
axs[1].set_ylabel("% Win Rate")
axs[1].set_title("CoNa")
# 添加颜色的含义图例 color for Win, Tie, Lose
axs[1].legend(loc="upper right")

plt.show()

# save high-quality figure
fig.savefig("HHH_score.png", dpi=300)
```
